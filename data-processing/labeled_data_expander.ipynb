# Notebook Description:

This notebooks reads in the data from the `labeled_data.zip` csv and maps each individual entry to multiple entries where there is only one parent url and one child url for each row. 

Essentially, it takes the list of child urls and makes one entry for each of those. 

```python
import pdb
import pandas as pd
```

```python
compact_df = pd.read_csv("../data/raw/labeled_data.zip")
```

```python
def parse_links_str(links):
    link_list = links.split("[")[1].split("]")[0].split(",")
    return [link.split("'")[-2] for link in link_list]
```

```python

expanded = [] # where we will store the transformed data
error_count = 0
for i in range(compact_df.shape[0]):
    title = compact_df.iloc[i].title
    base_url = compact_df.iloc[i].url
    try:
#         pass to external function to break the single string into a list of urls
        links = parse_links_str(compact_df.iloc[i,2])
    except: 
        error_count += 1
        continue
    
    
    for link in links:
        link_title = link.split("/")[-1]
        expanded.append((title, base_url, link_title, link))
        
print("Num Errors:", error_count)
expanded = pd.DataFrame(expanded, columns=["parent_title", "parent_url", "child_title", "child_url" ])

    
    
    
```

```python
expanded.to_csv("../data/raw/expanded_labeled_data.csv", index=False)
```
